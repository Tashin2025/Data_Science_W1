```python
n, p, m = 3, 2, 4

A = np.random.rand(n, p)
B = np.random.rand(p, m)

dot_product = np.dot(A, B)

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Dot Product (A × B):\n", dot_product)
```
